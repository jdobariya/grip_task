#### Task 4: To explore Business Analytics
In the given task we perform ‘Exploratory Data Analysis’ on the provided dataset ‘SampleSuperstore’ to provide differnet insights on the business.

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SampleSuperStore.csv')
```

```python
df.head(5)
```

```python
df.describe()
```

```python
corr = df.corr().iloc[1:,1:]
with sns.axes_style("white"):
    sns.heatmap(corr, cmap="YlGnBu",annot=True)
    plt.show()
```

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(10,9))
sns.set(style="darkgrid")
ax = sns.countplot(x="Ship Mode",hue='Region', data=df,palette="Set1")
ax.set_title('Ship Mode Frequency in Different Region of Country U.S')
```

```python
plt.figure(figsize=(10,9))
sns.set(style="whitegrid")
ax = sns.countplot(x="Segment",hue='Region', data=df,palette='viridis')
ax.set_title('Segemnt in Diff. Region of Country U.S')
```

```python
sales = df.groupby('Region')['Sales'].sum()
fig, ax1 = plt.subplots(figsize=(9, 7))
ax1.bar(x=sales.index,height=sales.values,color=['tomato','cyan','lightgreen','gray'])
ax1.set_xlabel('Regions')
ax1.set_ylabel('Total Sales')
ax1.set_title('Total Sales in Differnt Region of U.S')
```

```python
sns.set(style="white")
sales = df.groupby('Sub-Category')['Sales'].sum()
fig, ax2 = plt.subplots(figsize=(9, 7))
ax2.bar(x=sales.index,height=sales.values,color=['#348781','#307D7E'])
ax2.set_xlabel('Sub-Category')
plt.xticks(rotation=70)
ax2.set_ylabel('Total Sales')
ax2.set_title('Total Sales by Sub-Category')
```

```python
sns.set(style="white")
sales = df.groupby('Sub-Category')['Profit'].sum()
fig, ax3 = plt.subplots(figsize=(9, 7))
ax3.bar(x=sales.index,height=sales.values,color=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'])
ax3.set_xlabel('Sub-Category')
plt.xticks(rotation=70)
ax3.set_ylabel('Total Profit')
ax3.set_title('Total Profit by Sub-Category')
```

```python
sales = df.groupby('Category')['Sales'].sum()
fig, ax4 = plt.subplots(figsize=(9, 7))
ax4.bar(x=sales.index,height=sales.values,color=['darkblue','orange','purple'])
ax4.set_xlabel('Category')
ax4.set_ylabel('Total Sales')
ax4.set_title('Total Sales in by Different Category of country U.S')
```

```python
plt.figure(figsize = (10, 9))
sns.countplot(df['Sub-Category']) 
plt.xticks(rotation = 50, ha = 'right', fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Sub-Category', fontsize = 15)
plt.ylabel('Frequency', fontsize = 15);
```

```python
labels = df['Category'].value_counts().index
sizes = df['Category'].value_counts().values
colors = ['yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0,0,0)
plt.figure(figsize=(7,5))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=260)
plt.title('Usage by Category')
plt.axis('equal')
plt.show()
```

```python
labels = df.groupby('Region')['Profit'].sum().index
sizes = df.groupby('Region')['Profit'].sum().values
colors = ['gold','yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0,0,0,0.1)
plt.figure(figsize=(7,5))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=260)
plt.title('Profit in Different U.S Region')
plt.axis('equal')
plt.show()
```

```python
labels = df.groupby('Category')['Profit'].sum().index
sizes = df.groupby('Category')['Profit'].sum().values
colors = ['gold', 'lightcoral', 'lightskyblue']
explode = (0,0,0.01)
plt.figure(figsize=(7,5))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=260)
plt.title('Profit by Different Category')
plt.axis('equal')
plt.show()
```

```python
state_sale = df.groupby('State')['Sales'].sum()
f, ax = plt.subplots(figsize=(6, 15))
ax = sns.barplot(x=sales.values, y=state_sale.index)
ax.set_title('Total Sales in Different State of U.S')
ax.set_xlabel('Total Sales')
ax.set_ylabel('States')
```

```python
state_sale = df.groupby('State')['Profit'].sum()
f, ax = plt.subplots(figsize=(6, 15))
ax = sns.barplot(x=sales.values, y=state_sale.index)
ax.set_title('Profit made in Different State of U.S')
ax.set_xlabel('Profit')
ax.set_ylabel('States')
```
